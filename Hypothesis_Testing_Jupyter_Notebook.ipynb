```python
#Null Hypothesis no difference in weight between gentoo and chinstrap penguins 
```

```python
#Alternate Hypothesis there is a difference in weight between gentoo and chinstrap penguins
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpl
from scipy import stats
```

```python
penguins = pd.read_csv(r"C:\Users\Jethro\OneDrive\Documents\Jethro's Folder\Data Analyst Related\Dataset\penguins_size.csv")
```

```python
penguins.describe()
```

```python
penguins.head()
penguins.dropna()
```

```python
penguins.groupby('species')["body_mass_g"].mean()
```

```python
filter_gentoo = penguins[penguins['species'] == "Gentoo"]
filter_gentoo.dropna()
```

```python
filter_chinstrap = penguins[penguins['species'] == "Chinstrap"]
filter_chinstrap.dropna()
```

```python
sampled_gentoo = filter_gentoo.sample(n =50, replace = True, random_state = 60)
sampled_gentoo.describe()
```

```python
sampled_chinstrap= filter_chinstrap.sample(n =50, replace = True, random_state = 60)
sampled_chinstrap.describe()
```

```python
stats.ttest_ind(a=sampled_gentoo['body_mass_g'], b=sampled_chinstrap['body_mass_g'], equal_var=False)
```

```python
#P-value is greater than SE; fail to reject the null hypothesis
```
